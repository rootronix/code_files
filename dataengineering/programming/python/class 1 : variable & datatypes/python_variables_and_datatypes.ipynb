<div style="border-left:4px solid #3B82F6;padding-left:12px;color:#1E293B;"><span style="font-size:18px;font-weight:700;">A. Understanding Variables</span></div>

#### 1. What is a Variable?

**What it is** — A name that *refers* to an object in memory. Python variables hold references, not the values themselves.

**Why it matters** — It explains Python's entire behaviour around copying, mutation and identity.

<div style="background:#EFF6FF;border-left:5px solid #3B82F6;padding:10px;border-radius:6px;color:#1E293B;"><b>Executive Insight</b><br>
No declaration needed: a name springs into existence the moment you assign to it.
</div>

<div style="background:#ECFDF5;border-left:5px solid #10B981;padding:8px 10px;border-radius:6px;color:#1E293B;"><b>&#127919; Key Takeaway</b> &nbsp;A variable is a label on an object &mdash; not a box that contains it.
</div>

```python
name = "John"
age = 25

print(name)
print(age)
```

#### 2. Creating Variables & Assignment ( = )

**What it is** — The <code>=</code> operator binds a name to an object; the variable is created automatically on first assignment.

**Why it matters** — It is the single most common operation you will write in Python.

<div style="background:#ECFDF5;border-left:5px solid #10B981;padding:8px 10px;border-radius:6px;color:#1E293B;"><b>&#127919; Key Takeaway</b> &nbsp;Assignment is name &rarr; object binding, not value storage.
</div>

```python
city = "Bengaluru"
x = 100

print(city)
print(x)
```

#### 3. Dynamic Typing

**What it is** — A name may reference objects of different types at different times; the type lives on the object, not the name.

**Why it matters** — Lets you write flexible code, but means types can silently change.

<div style="background:#F8FAFC;border-left:5px solid #64748B;padding:10px;border-radius:6px;color:#1E293B;"><b>Important Note</b><br>
Flexibility is powerful but can hide bugs &mdash; keep one name to one role.
</div>

<div style="background:#ECFDF5;border-left:5px solid #10B981;padding:8px 10px;border-radius:6px;color:#1E293B;"><b>&#127919; Key Takeaway</b> &nbsp;Variables are untyped; objects are typed.
</div>

```python
data = "Python"
print(data)
print(type(data))

```

<div style="border-left:4px solid #3B82F6;padding-left:12px;color:#1E293B;"><span style="font-size:18px;font-weight:700;">B. Assignment Techniques</span></div>

#### 1. Reassignment

**What it is** — Pointing an existing name at a new object.

**Why it matters** — State changes throughout a program are just reassignments.

<div style="background:#ECFDF5;border-left:5px solid #10B981;padding:8px 10px;border-radius:6px;color:#1E293B;"><b>&#127919; Key Takeaway</b> &nbsp;Reassigning rebinds the name; the old object is left for cleanup.
</div>

```python
score = 50
score = 100
print(score)
```

#### 2. Multiple & Chain Assignment

**What it is** — Bind several names in one line (<code>x, y = 1, 2</code>) or bind many names to one object (<code>a = b = c = 5</code>).

**Why it matters** — Cleaner setup code and tidy initialisation.

<div style="background:#ECFDF5;border-left:5px solid #10B981;padding:8px 10px;border-radius:6px;color:#1E293B;"><b>&#127919; Key Takeaway</b> &nbsp;Multiple assignment unpacks; chain assignment shares one object.
</div>

```python
x, y, z = 10, 20, 30
print(x, y, z)

a = b = c = 5
print(a, b, c)
```

#### 3. Variable Swapping

**What it is** — Exchange two values in one line with no temporary variable.

**Why it matters** — A clean, Pythonic idiom you will use constantly.

<div style="background:#ECFDF5;border-left:5px solid #10B981;padding:8px 10px;border-radius:6px;color:#1E293B;"><b>&#127919; Key Takeaway</b> &nbsp;<code>x, y = y, x</code> swaps in place &mdash; no temp needed.
</div>

```python
x = 100
y = 200

x, y = y, x
print(x, y)
```

#### 4. Augmented Assignment ( += )

**What it is** — <code>x += 1</code> is shorthand, but for <i>mutable</i> objects it modifies in place, while <code>x = x + 1</code> always rebinds.

**Why it matters** — The difference quietly causes shared-list bugs.

<div style="background:#EFF6FF;border-left:5px solid #3B82F6;padding:10px;border-radius:6px;color:#1E293B;"><b>Executive Insight</b><br>
Watch the <code>id()</code> below: the list keeps its identity after <code>+=</code>, but the integer gets a brand-new one.
</div>

<div style="background:#ECFDF5;border-left:5px solid #10B981;padding:8px 10px;border-radius:6px;color:#1E293B;"><b>&#127919; Key Takeaway</b> &nbsp;For immutables they match; for lists, <code>+=</code> mutates the same object.
</div>

```python
n = 5
print(n)
n += 2
print(n)
```

#### 5. Deleting Variables ( del )

**What it is** — <code>del</code> removes the name from the namespace (it does not directly destroy the object).

**Why it matters** — Frees names and lets the object be garbage-collected when unreferenced.

<div style="background:#ECFDF5;border-left:5px solid #10B981;padding:8px 10px;border-radius:6px;color:#1E293B;"><b>&#127919; Key Takeaway</b> &nbsp;<code>del</code> unbinds a name; using it afterwards raises <code>NameError</code>.
</div>

```python
name = "Python"
del name

try:
    print(name)
except NameError as e:
    print("NameError:", e)
```

<div style="border-left:4px solid #3B82F6;padding-left:12px;color:#1E293B;"><span style="font-size:18px;font-weight:700;">C. Identifiers & Keywords</span></div>

#### 1. Naming Rules & Best Practices

**What it is** — Identifiers may use letters, digits and underscores, cannot start with a digit, and are case-sensitive.

**Why it matters** — Good names are the cheapest documentation you will ever write.

<div style="background:#ECFDF5;border-left:5px solid #10B981;padding:10px;border-radius:6px;color:#1E293B;"><b>Best Practice</b><br>
<code>total_revenue</code> beats <code>tr</code>; <code>customer_count</code> beats <code>c</code>.
</div>

<div style="background:#ECFDF5;border-left:5px solid #10B981;padding:8px 10px;border-radius:6px;color:#1E293B;"><b>&#127919; Key Takeaway</b> &nbsp;Use descriptive <code>snake_case</code>; avoid single letters except for counters.
</div>

```python
customer_name1 = "John"
customer_name2 = "john"
Customer_name2 = "Peter"

print(customer_name1,customer_name2, Customer_name2)
print("Case-sensitive:", customer_name1==customer_name2)
```

#### 2. Reserved Keywords

**What it is** — Words reserved by the language (e.g. <code>for</code>, <code>class</code>, <code>None</code>) that cannot be used as names.

**Why it matters** — Using one as a variable is an instant <code>SyntaxError</code>.

<div style="background:#ECFDF5;border-left:5px solid #10B981;padding:8px 10px;border-radius:6px;color:#1E293B;"><b>&#127919; Key Takeaway</b> &nbsp;Check with the <code>keyword</code> module instead of memorising the list.
</div>

```python
import keyword

print("Is 'import' a keyword?", keyword.iskeyword("import"))
print("Is 'data' a keyword?", keyword.iskeyword("data"))
```

<div style="background:linear-gradient(90deg,#EFF6FF,#ECFDF5);border:1px solid #CBD5E1;border-radius:10px;padding:16px 20px;color:#1E293B;"><div style="font-size:13px;letter-spacing:2px;color:#3B82F6;font-weight:700;">SECTION II</div><div style="font-size:22px;font-weight:800;margin-top:2px;">The Python Object Model</div><div style="font-size:14px;color:#64748B;margin-top:6px;">In Python, everything is an object &mdash; and names are just references to them.</div></div>

<div style="border-left:4px solid #3B82F6;padding-left:12px;color:#1E293B;"><span style="font-size:18px;font-weight:700;">A. Objects & References</span></div>

#### 1. Everything is an Object

**What it is** — Integers, strings, functions and classes are all objects with a type.

**Why it matters** — One consistent mental model explains the whole language.

<div style="background:#ECFDF5;border-left:5px solid #10B981;padding:8px 10px;border-radius:6px;color:#1E293B;"><b>&#127919; Key Takeaway</b> &nbsp;If it exists in Python, it is an object.
</div>

```python
x = 10
name = "Python"

print(type(x))
print(type(name))
print(type(print))
```

#### 2. Identity, Type & Value

**What it is** — Every object carries an identity (<code>id()</code>), a type (<code>type()</code>) and a value.

**Why it matters** — These three properties define what an object <i>is</i>.

<div style="background:#F8FAFC;border-left:5px solid #64748B;padding:10px;border-radius:6px;color:#1E293B;"><b>Important Note</b><br>
<code>id()</code> values vary by machine and run &mdash; only equality between ids is meaningful.
</div>

<div style="background:#ECFDF5;border-left:5px solid #10B981;padding:8px 10px;border-radius:6px;color:#1E293B;"><b>&#127919; Key Takeaway</b> &nbsp;Identity = address, Type = kind, Value = content.
</div>

```python
x = 100
print("id  :", id(x))
print("type:", type(x))
print("val :", x)
```

#### 3. Variables as References

**What it is** — Two names assigned to the same object share one identity until one is rebound.

**Why it matters** — This is why mutating through one name is visible through the other.

<div style="background:#EFF6FF;border-left:5px solid #3B82F6;padding:10px;border-radius:6px;color:#1E293B;"><b>Executive Insight</b><br>
<code>is</code> compares identity; <code>==</code> compares value. They are not the same test.
</div>

<div style="background:#ECFDF5;border-left:5px solid #10B981;padding:8px 10px;border-radius:6px;color:#1E293B;"><b>&#127919; Key Takeaway</b> &nbsp;<code>y = x</code> copies the reference, not the object.
</div>

```python
x = 'a'
y = x
print(id(x) == id(y))   # same object
print(x is y)
```

<div style="border-left:4px solid #3B82F6;padding-left:12px;color:#1E293B;"><span style="font-size:18px;font-weight:700;">B. Introspection Tools</span></div>

#### 1. type(), id(), dir(), help()

**What it is** — The core inspection toolkit: kind, identity, available members, and live documentation.

**Why it matters** — They let you investigate any object without leaving the REPL.

<div style="background:#ECFDF5;border-left:5px solid #10B981;padding:8px 10px;border-radius:6px;color:#1E293B;"><b>&#127919; Key Takeaway</b> &nbsp;<code>type</code>=what, <code>id</code>=where, <code>dir</code>=what-it-can-do, <code>help</code>=how.
</div>

```python
s = "Python"
print(type(s))
print("methods sample:", [m for m in dir(s) if not m.startswith("_")][:5])
print("doc of len():", len.__doc__)
```

<div style="border-left:4px solid #3B82F6;padding-left:12px;color:#1E293B;"><span style="font-size:18px;font-weight:700;">C. Object Representation</span></div>

#### 1. str() vs repr()

**What it is** — <code>str()</code> gives a readable form for users; <code>repr()</code> gives an unambiguous form for developers.

**Why it matters** — Logging and debugging rely on <code>repr()</code> showing the real structure.

<div style="background:#ECFDF5;border-left:5px solid #10B981;padding:8px 10px;border-radius:6px;color:#1E293B;"><b>&#127919; Key Takeaway</b> &nbsp;<code>str</code> = friendly; <code>repr</code> = precise (note the quotes and escapes).
</div>

```python
s = "line1\nline2"
print("str :", str(s))
print("repr:", repr(s))
```

<div style="background:linear-gradient(90deg,#EFF6FF,#ECFDF5);border:1px solid #CBD5E1;border-radius:10px;padding:16px 20px;color:#1E293B;"><div style="font-size:13px;letter-spacing:2px;color:#3B82F6;font-weight:700;">SECTION III</div><div style="font-size:22px;font-weight:800;margin-top:2px;">Memory Management</div><div style="font-size:14px;color:#64748B;margin-top:6px;">Where objects live, how Python tracks them, and when it reclaims them.</div></div>

<div style="border-left:4px solid #3B82F6;padding-left:12px;color:#1E293B;"><span style="font-size:18px;font-weight:700;">A. Storage Model</span></div>

#### 1. References & Object Size

**What it is** — A name holds a reference; <code>sys.getsizeof()</code> reports an object's memory footprint in bytes.

**Why it matters** — Useful when profiling memory-heavy data.

<div style="background:#F8FAFC;border-left:5px solid #64748B;padding:10px;border-radius:6px;color:#1E293B;"><b>Important Note</b><br>
Exact byte counts depend on your Python build and platform.
</div>

<div style="background:#ECFDF5;border-left:5px solid #10B981;padding:8px 10px;border-radius:6px;color:#1E293B;"><b>&#127919; Key Takeaway</b> &nbsp;Bigger or richer objects cost more bytes &mdash; measure, don't guess.
</div>

```python
import sys
x=100
print("int  :", sys.getsizeof(x), "bytes")
print("str  :", sys.getsizeof("Python"), "bytes")
print("list :", sys.getsizeof([1, 2, 3]), "bytes")
```

#### 2. Stack vs Heap (Conceptual)

**What it is** — References conceptually live on the stack; the actual objects live on the heap.

**Why it matters** — Explains why assigning a name is cheap even for huge objects.

<div style="background:#ECFDF5;border-left:5px solid #10B981;padding:8px 10px;border-radius:6px;color:#1E293B;"><b>&#127919; Key Takeaway</b> &nbsp;Names are lightweight pointers; the data sits elsewhere on the heap.
</div>

```python
x = 10           # name "x" (stack) -> int object 10 (heap)
y = x            # second pointer to the SAME heap object
print(id(x) == id(y))
```

<div style="border-left:4px solid #3B82F6;padding-left:12px;color:#1E293B;"><span style="font-size:18px;font-weight:700;">B. Object Lifecycle</span></div>

#### 1. Reference Counting

**What it is** — CPython keeps a count of references to each object via <code>sys.getrefcount()</code>.

**Why it matters** — When the count hits zero the object can be freed.

<div style="background:#F8FAFC;border-left:5px solid #64748B;padding:10px;border-radius:6px;color:#1E293B;"><b>Important Note</b><br>
The count looks one higher than expected because the argument itself is a temporary reference.
</div>

<div style="background:#ECFDF5;border-left:5px solid #10B981;padding:8px 10px;border-radius:6px;color:#1E293B;"><b>&#127919; Key Takeaway</b> &nbsp;More references = the object stays alive longer.
</div>

```python
import sys
obj = []
print("refs:", sys.getrefcount(obj))
alias = obj
print("refs after alias:", sys.getrefcount(obj))
```

#### 2. Garbage Collection

**What it is** — Python automatically reclaims objects with no remaining references.

**Why it matters** — You manage logic; the runtime manages memory.

<div style="background:#ECFDF5;border-left:5px solid #10B981;padding:8px 10px;border-radius:6px;color:#1E293B;"><b>&#127919; Key Takeaway</b> &nbsp;Drop the last reference (e.g. with <code>del</code>) and cleanup is automatic.
</div>

```python
import gc
x = [1, 2, 3]
del x                 # last reference gone
collected = gc.collect()
print("GC ran, objects collected:", collected)
```

<div style="border-left:4px solid #3B82F6;padding-left:12px;color:#1E293B;"><span style="font-size:18px;font-weight:700;">A. Integers (int)</span></div>

#### 1. Numeric Literals: Decimal, Binary, Octal, Hex

**What it is** — The same integer can be written in base 10, base 2 (<code>0b</code>), base 8 (<code>0o</code>) or base 16 (<code>0x</code>).

**Why it matters** — Bit masks, permissions and colours are far clearer in their natural base.

<div style="background:#ECFDF5;border-left:5px solid #10B981;padding:8px 10px;border-radius:6px;color:#1E293B;"><b>&#127919; Key Takeaway</b> &nbsp;Different notations, one identical integer value.
</div>

```python
dec=255
print(dec)        # decimal
bin=0b11111111
print(bin) # binary
oct=0o377
print(oct)      # octal
hex=0x7b
print(0xFF)       # hex

print(dec == bin == oct)
```

#### 2. Underscore Separators

**What it is** — Underscores group digits in large numeric literals; they are ignored by Python.

**Why it matters** — Readability for big numbers, with zero runtime cost.

<div style="background:#ECFDF5;border-left:5px solid #10B981;padding:8px 10px;border-radius:6px;color:#1E293B;"><b>&#127919; Key Takeaway</b> &nbsp;<code>1_500_000</code> reads instantly; the value is plain <code>1500000</code>.
</div>

```python
population = 1_500_000
print(population)
print(population == 1500000)
```

<div style="border-left:4px solid #3B82F6;padding-left:12px;color:#1E293B;"><span style="font-size:18px;font-weight:700;">B. Floating-Point (float)</span></div>

#### 1. Float Basics

**What it is** — Numbers with a decimal point or exponent; division always yields a float.

**Why it matters** — Most measured, scientific and financial values are floats.

<div style="background:#F8FAFC;border-left:5px solid #64748B;padding:10px;border-radius:6px;color:#1E293B;"><b>Important Note</b><br>
<code>0.1 + 0.2</code> is famously not exactly <code>0.3</code> &mdash; binary floats can't store every decimal.
</div>

<div style="background:#ECFDF5;border-left:5px solid #10B981;padding:8px 10px;border-radius:6px;color:#1E293B;"><b>&#127919; Key Takeaway</b> &nbsp;Use floats for fractions &mdash; but expect tiny rounding error.
</div>

```python
print(type(3.14))
print(10 / 3)
print(10 / 2)
```

#### 2. Special Float Values: inf & nan

**What it is** — Floats include positive/negative infinity and NaN (Not a Number).

**Why it matters** — They appear in overflow, undefined math and missing data.

<div style="background:#ECFDF5;border-left:5px solid #10B981;padding:8px 10px;border-radius:6px;color:#1E293B;"><b>&#127919; Key Takeaway</b> &nbsp;<code>nan</code> is never equal to anything &mdash; even itself.
</div>

```python
pos = float('inf')
neg = float('-inf')
nan = float('nan')
print(pos, neg, nan)
print('inf > huge :', pos > 1e308)
print('nan == nan :', nan == nan)
```

<div style="border-left:4px solid #3B82F6;padding-left:12px;color:#1E293B;"><span style="font-size:18px;font-weight:700;">A. String Literals</span></div>

#### 1. Quote Variants & f-strings

**What it is** — Use single, double or triple quotes; prefix <code>r</code> for raw (no escapes) or <code>f</code> for formatted strings.

**Why it matters** — The right quote style avoids escaping headaches and powers clean formatting.

<div style="background:#ECFDF5;border-left:5px solid #10B981;padding:8px 10px;border-radius:6px;color:#1E293B;"><b>&#127919; Key Takeaway</b> &nbsp;Triple quotes span lines; <code>r"..."</code> keeps backslashes literal; <code>f"..."</code> embeds expressions.
</div>

```python
single = 'hi'
double = "hi"
triple = """line A
line B"""
print(single, double)
print(triple)
```

<div style="border-left:4px solid #3B82F6;padding-left:12px;color:#1E293B;"><span style="font-size:18px;font-weight:700;">B. Escape Sequences</span></div>

#### 1. Common Escape Sequences

**What it is** — Backslash codes insert special characters: <code>\n \t \r \b \\ \' \"</code>.

**Why it matters** — They control layout and let you embed quotes safely.

<div style="background:#ECFDF5;border-left:5px solid #10B981;padding:8px 10px;border-radius:6px;color:#1E293B;"><b>&#127919; Key Takeaway</b> &nbsp;Most courses teach only <code>\n</code> &mdash; know the full set.
</div>

```python
raw = r"C:\new\path"
print(raw)
name = "Arun"
fstr = f"Hello {name}, score={40+2}"
print(fstr)
```

```python
print("a\tb")        # tab
print("line1\nline2") # newline
print("quote: \"hi\"") # literal quotes
print("back\\slash")  # literal backslash
```

<div style="border-left:4px solid #3B82F6;padding-left:12px;color:#1E293B;"><span style="font-size:18px;font-weight:700;">C. String Behaviour</span></div>

#### 1. String Comparison

**What it is** — Strings compare lexicographically by Unicode code point, and comparison is case-sensitive.

**Why it matters** — Sorting and searching depend on these exact rules.

<div style="background:#ECFDF5;border-left:5px solid #10B981;padding:8px 10px;border-radius:6px;color:#1E293B;"><b>&#127919; Key Takeaway</b> &nbsp;Uppercase letters sort before lowercase because their code points are lower.
</div>

```python
print("apple" < "banana")    # b > a
print("Apple" < "apple")     # 'A'(65) < 'a'(97)
print("apple" == "Apple")    # case-sensitive
```

#### 2. Character Encoding: ord() & chr()

**What it is** — <code>ord()</code> maps a character to its Unicode code point; <code>chr()</code> maps a code point back.

**Why it matters** — Foundational for encryption, hashing and text processing.

<div style="background:#ECFDF5;border-left:5px solid #10B981;padding:8px 10px;border-radius:6px;color:#1E293B;"><b>&#127919; Key Takeaway</b> &nbsp;They are exact inverses: <code>chr(ord(c)) == c</code>.
</div>

```python
print(ord('A'))
print(chr(65))
print(chr(ord('A') + 1))   # next letter
```

#### 3. String Interning (Pooling)

**What it is** — Python reuses one object for identical short string literals, so they share identity.

**Why it matters** — Saves memory and makes some equality checks faster.

<div style="background:#F8FAFC;border-left:5px solid #64748B;padding:10px;border-radius:6px;color:#1E293B;"><b>Important Note</b><br>
Interning is an optimisation, not a guarantee &mdash; never use <code>is</code> to compare string values.
</div>

<div style="background:#ECFDF5;border-left:5px solid #10B981;padding:8px 10px;border-radius:6px;color:#1E293B;"><b>&#127919; Key Takeaway</b> &nbsp;Equal literals may be the <i>same</i> object &mdash; but rely on <code>==</code>, not <code>is</code>, for value checks.
</div>

```python
a = "python"
b = "python"
print("same object:", a is b)
print("equal value :", a == b)
```

<div style="background:linear-gradient(90deg,#EFF6FF,#ECFDF5);border:1px solid #CBD5E1;border-radius:10px;padding:16px 20px;color:#1E293B;"><div style="font-size:13px;letter-spacing:2px;color:#3B82F6;font-weight:700;">SECTION VI</div><div style="font-size:22px;font-weight:800;margin-top:2px;">Booleans &amp; None</div><div style="font-size:14px;color:#64748B;margin-top:6px;">Truthiness, boolean conversion, and the classic None / False / 0 confusion.</div></div>

<div style="border-left:4px solid #3B82F6;padding-left:12px;color:#1E293B;"><span style="font-size:18px;font-weight:700;">A. Boolean Logic</span></div>

#### 1. Truth Value Testing & bool()

**What it is** — Every object is truthy or falsy. Falsy values include <code>0</code>, <code>""</code>, empty containers and <code>None</code>.

**Why it matters** — <code>if</code> statements test truthiness, not just <code>True</code>/<code>False</code>.

<div style="background:#ECFDF5;border-left:5px solid #10B981;padding:8px 10px;border-radius:6px;color:#1E293B;"><b>&#127919; Key Takeaway</b> &nbsp;Empty / zero / None &rarr; falsy; almost everything else &rarr; truthy.
</div>

```python
print(bool(""))      # empty string
print(bool("Python")) # non-empty
print(bool(0))
print(bool(-1))      # any non-zero is True
print(bool([]))      # empty list
print(bool(None))
```

<div style="border-left:4px solid #3B82F6;padding-left:12px;color:#1E293B;"><span style="font-size:18px;font-weight:700;">B. None vs False vs 0</span></div>

#### 1. None vs False vs 0

**What it is** — <code>None</code> means 'no value'; <code>False</code> is a boolean; <code>0</code> is a number. <code>False</code> equals <code>0</code> by value, but <code>None</code> equals neither.

**Why it matters** — Confusing them produces subtle logic bugs.

<div style="background:#EFF6FF;border-left:5px solid #3B82F6;padding:10px;border-radius:6px;color:#1E293B;"><b>Executive Insight</b><br>
<code>False == 0</code> is <code>True</code> because <code>bool</code> is a subclass of <code>int</code>.
</div>

<div style="background:#ECFDF5;border-left:5px solid #10B981;padding:8px 10px;border-radius:6px;color:#1E293B;"><b>&#127919; Key Takeaway</b> &nbsp;Test for <code>None</code> with <code>is None</code>, never with <code>== False</code>.
</div>

```python
print("None == False:", None == False)
print("None == 0    :", None == 0)
print("False == 0   :", False == 0)
print("None is None :", None is None)
```

<div style="background:linear-gradient(90deg,#EFF6FF,#ECFDF5);border:1px solid #CBD5E1;border-radius:10px;padding:16px 20px;color:#1E293B;"><div style="font-size:13px;letter-spacing:2px;color:#3B82F6;font-weight:700;">SECTION VII</div><div style="font-size:22px;font-weight:800;margin-top:2px;">Type Conversion &amp; Identification</div><div style="font-size:14px;color:#64748B;margin-top:6px;">Convert safely between types, check types correctly, and validate user input.</div></div>

<div style="border-left:4px solid #3B82F6;padding-left:12px;color:#1E293B;"><span style="font-size:18px;font-weight:700;">A. Type Conversion</span></div>

#### 1. Core Conversions

**What it is** — <code>int()</code>, <code>float()</code> and <code>str()</code> convert between types; <code>int()</code> truncates toward zero.

**Why it matters** — User input and file data arrive as strings and must be converted.

<div style="background:#ECFDF5;border-left:5px solid #10B981;padding:8px 10px;border-radius:6px;color:#1E293B;"><b>&#127919; Key Takeaway</b> &nbsp;Convert explicitly &mdash; Python won't mix strings and numbers for you.
</div>

```python
print(int("10") + 5)
print(float("3"))
print(int(3.9))        # truncates, not rounds
print(str(100) + "%")

print("10"+"5")
```

#### 2. Conversion Edge Cases

**What it is** — Booleans are integers under the hood, so <code>int(True)</code> is 1 and <code>bool(0)</code> is False.

**Why it matters** — These are classic interview questions and real bug sources.

<div style="background:#ECFDF5;border-left:5px solid #10B981;padding:8px 10px;border-radius:6px;color:#1E293B;"><b>&#127919; Key Takeaway</b> &nbsp;<code>True</code> behaves as <code>1</code> and <code>False</code> as <code>0</code> in arithmetic.
</div>

```python
print(int(True))    # 1
print(float(False)) # 0.0
print(bool(0))      # False
print(bool(-1))     # True
print(True + True)  # 2
```

<div style="border-left:4px solid #3B82F6;padding-left:12px;color:#1E293B;"><span style="font-size:18px;font-weight:700;">B. Type Identification</span></div>

#### 1. type() vs isinstance()

**What it is** — <code>type()</code> checks for an exact type; <code>isinstance()</code> also accepts subclasses.

**Why it matters** — Choosing wrong breaks code that relies on inheritance.

<div style="background:#ECFDF5;border-left:5px solid #10B981;padding:10px;border-radius:6px;color:#1E293B;"><b>Best Practice</b><br>
Use <code>isinstance(x, int)</code> in real code; reserve <code>type(x) is int</code> for strict checks.
</div>

<div style="background:#ECFDF5;border-left:5px solid #10B981;padding:8px 10px;border-radius:6px;color:#1E293B;"><b>&#127919; Key Takeaway</b> &nbsp;Prefer <code>isinstance()</code> for checks &mdash; note <code>True</code> is an <code>int</code>.
</div>

```python
print(isinstance(10, int))
print(isinstance(True, int))   # True! bool subclasses int
print(type(True) is int)       # False (exact type is bool)
print(isinstance("hi", (int, str)))
```

<div style="border-left:4px solid #3B82F6;padding-left:12px;color:#1E293B;"><span style="font-size:18px;font-weight:700;">C. Safe Input Handling</span></div>

#### 1. Validate Before Converting

**What it is** — Check that text is numeric (e.g. <code>str.isdigit()</code>) before calling <code>int()</code> to avoid crashes.

**Why it matters** — Unvalidated <code>int(input())</code> raises <code>ValueError</code> on bad input.

<div style="background:#F8FAFC;border-left:5px solid #64748B;padding:10px;border-radius:6px;color:#1E293B;"><b>Important Note</b><br>
We simulate <code>input()</code> with fixed strings here so the cell runs without prompting.
</div>

<div style="background:#ECFDF5;border-left:5px solid #10B981;padding:8px 10px;border-radius:6px;color:#1E293B;"><b>&#127919; Key Takeaway</b> &nbsp;Validate first, convert second &mdash; never trust raw input.
</div>

```python
def to_age(raw):
    if raw.isdigit():
        return int(raw)
    return f"invalid: {raw!r}"

print(to_age("25"))
print(to_age("2a"))
```

<div style="background:linear-gradient(90deg,#EFF6FF,#ECFDF5);border:1px solid #CBD5E1;border-radius:10px;padding:16px 20px;color:#1E293B;"><div style="font-size:13px;letter-spacing:2px;color:#3B82F6;font-weight:700;">SECTION VIII</div><div style="font-size:22px;font-weight:800;margin-top:2px;">Hashability</div><div style="font-size:14px;color:#64748B;margin-top:6px;">The most-skipped data-type concept &mdash; and the key to sets and dict keys.</div></div>

<div style="border-left:4px solid #3B82F6;padding-left:12px;color:#1E293B;"><span style="font-size:18px;font-weight:700;">A. Hashing</span></div>

#### 1. hash(): Hashable vs Unhashable

**What it is** — Hashing maps an object to an integer. Immutable objects (numbers, strings, tuples) are hashable; mutable ones (lists, dicts) are not.

**Why it matters** — Only hashable objects can be dict keys or set members.

<div style="background:#EFF6FF;border-left:5px solid #3B82F6;padding:10px;border-radius:6px;color:#1E293B;"><b>Executive Insight</b><br>
Hashability tracks immutability: mutate-able means un-hashable, because the hash would change underneath you.
</div>

<div style="background:#ECFDF5;border-left:5px solid #10B981;padding:8px 10px;border-radius:6px;color:#1E293B;"><b>&#127919; Key Takeaway</b> &nbsp;If you can't <code>hash()</code> it, you can't use it as a dict key.
</div>

```python
print("hash str  :", hash("Python"))   # works (value varies)
print("hash tuple:", hash((1, 2, 3)))       # works

try:
    hash([1, 2, 3])                          # list = unhashable
except TypeError as e:
    print("hash list :", "TypeError:", e)
```

<div style="background:linear-gradient(90deg,#EFF6FF,#ECFDF5);border:1px solid #CBD5E1;border-radius:10px;padding:16px 20px;color:#1E293B;"><div style="font-size:13px;letter-spacing:2px;color:#3B82F6;font-weight:700;">SECTION IX</div><div style="font-size:22px;font-weight:800;margin-top:2px;">Documentation</div><div style="font-size:14px;color:#64748B;margin-top:6px;">Make objects self-describing with docstrings.</div></div>

<div style="border-left:4px solid #3B82F6;padding-left:12px;color:#1E293B;"><span style="font-size:18px;font-weight:700;">A. Docstrings</span></div>

#### 1. Docstrings & __doc__

**What it is** — A triple-quoted string as the first statement of a function, class or module becomes its documentation.

**Why it matters** — It powers <code>help()</code>, IDE tooltips and auto-generated docs.

<div style="background:#ECFDF5;border-left:5px solid #10B981;padding:8px 10px;border-radius:6px;color:#1E293B;"><b>&#127919; Key Takeaway</b> &nbsp;Document intent at the source; read it back via <code>__doc__</code>.
</div>

```python
def to_celsius(f):
    """Convert Fahrenheit to Celsius."""
    return (f - 32) * 5 / 9

print(to_celsius(98.6))
print(to_celsius.__doc__)
```
